```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
cd = pd.read_csv('Company_data.csv')
cd.head(5)
```

```python
cd.describe().T
```

```python
cd.info()
```

```python
cd.isnull().sum()
```

```python
cd1 = pd.get_dummies(cd,columns=['Urban','US'])
cd1.head()
```

```python
cd1['ShelveLoc'] = cd1['ShelveLoc'].replace({'Bad': 1, 'Medium':2,'Good' :3})
cd1.head()
```

```python
cd1['Sales'] = cd1['Sales'].astype(int)
```

```python
X = cd1.drop('ShelveLoc',axis=1)
y= cd1['ShelveLoc']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.20, random_state=40)
```

```python
RFC = RandomForestClassifier(n_estimators=1000,max_depth=3, random_state=1)
```

```python
RFC.fit(X_train,y_train)
```

```python
pred=RFC.predict(X_test)
np.mean(pred==y_test)
```

```python
print(RFC.score(X_train, y_train)) 
print(RFC.score(X_test, y_test))
```
